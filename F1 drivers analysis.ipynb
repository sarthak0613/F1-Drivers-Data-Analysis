```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('F1Drivers_Dataset.csv')
```

```python
df.head(3)
```

```python
# Driver: The name of the Formula 1 driver.
# Nationality: The nationality of the driver.
# Seasons: The number of seasons the driver has competed in F1.
# Championships: The number of F1 World Championships the driver has won.
# Race_Entries: The total number of F1 race entries by the driver.
# Race_Starts: The total number of F1 race starts by the driver.
# Pole_Positions: The total number of times the driver has started a race from the pole position.
# Race_Wins: The total number of F1 race wins by the driver.
# Podiums: The total number of times the driver has finished on the podium (top 3) in F1 races.
# Fastest_Laps: The total number of fastest laps achieved by the driver in F1 races.
# Points: The total number of championship points scored by the driver in F1 races.
# Active: A binary indicator (0 or 1) representing whether the driver is currently active in F1.
# ChampionshipYears: The years in which the driver won F1 World Championships.
# Decade: The decade in which the driver's career primarily falls.
# Pole_Rate: The percentage of races in which the driver started from pole position.
# Start_Rate: The percentage of races in which the driver participated out of the total entries.
# Win_Rate: The percentage of races won by the driver out of total starts.
# Podium_Rate: The percentage of races in which the driver finished on the podium.
# FastLap_Rate: The percentage of races in which the driver achieved the fastest lap.
# Points_Per_Entry: The average number of championship points scored per race entry.
# Years_Active: The number of years the driver has been active in F1.
# Champion: A binary indicator (0 or 1) representing whether the driver has won at least one F1 World Championship.
```

```python
#Correlation matrix
```

```python
corr_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
corr_df = df[['Championships', 'Race_Starts', 'Pole_Positions', 'Race_Wins', 'Podiums', 'Fastest_Laps', 'Points', 'Years_Active']]
plt.figure(figsize = (15, 8))
mask = np.triu(np.ones_like(corr_df.corr(), dtype = bool))
heatmap = sns.heatmap(corr_df.corr(), cmap = 'crest', mask = mask, annot = True, linewidths = .5)
heatmap.set_title('Correlation Heatmap', fontdict = {'fontsize' : 15}, pad = 12)
plt.show()
```

```python
# Active vs. Non-active Drivers
#True=Active
#False=Non-Active
```

```python
active_counts = df['Active'].value_counts()
active_counts.plot(kind='pie', autopct='%1.1f%%', figsize=(6, 6))
plt.title('Active vs. Non-active Drivers')
plt.ylabel('')
plt.show()
```

```python
#Distribution of Nationalities
```

```python
nationality_counts = df['Nationality'].value_counts()
nationality_counts.plot(kind='bar', figsize=(10, 6))
plt.title('Distribution of Nationalities')
plt.xlabel('Nationality')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
#Top Nationalities with Most Race Wins
```

```python
top_nationalities = df.groupby('Nationality')['Race_Wins'].sum().nlargest(5)
top_nationalities.plot(kind='bar', figsize=(10, 6))
plt.title('Top 5 Nationalities with Most Race Wins')
plt.xlabel('Nationality')
plt.ylabel('Total Race Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
#Relationship between Pole Positions and Race Wins
```

```python
plt.scatter(df['Pole_Positions'], df['Race_Wins'], alpha=0.5)
plt.title('Pole Positions vs. Race Wins')
plt.xlabel('Pole Positions')
plt.ylabel('Race Wins')
plt.show()
```

```python
#Win Rate Comparison for Nationalities
```

```python
top_nationalities_win_rate = df.groupby('Nationality')['Win_Rate'].mean().nlargest(5)
top_nationalities_win_rate.plot(kind='bar', figsize=(10, 6))
plt.title('Top 5 Nationalities with the Highest Win Rates')
plt.xlabel('Nationality')
plt.ylabel('Average Win Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top Drivers with the Most Race Wins
```

```python
top_drivers_race_wins = df.nlargest(10, 'Race_Wins')
top_drivers_race_wins.plot(kind='bar', x='Driver', y='Race_Wins', figsize=(12, 6))
plt.title('Top 10 Drivers with the Most Race Wins')
plt.xlabel('Driver')
plt.ylabel('Race Wins')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 Active Drivers with the Highest Win Rates 
```

```python
top_active_drivers_win_rate = df[df['Active'] == 1].nlargest(10, 'Win_Rate')
top_active_drivers_win_rate.plot(kind='bar', x='Driver', y='Win_Rate', figsize=(12, 6))
plt.title('Top 10 Active Drivers with the Highest Win Rates')
plt.xlabel('Driver')
plt.ylabel('Win Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 Nationalities with the Highest Average Pole Rates
```

```python
avg_pole_rate_by_nationality = df.groupby('Nationality')['Pole_Rate'].mean().nlargest(10)
avg_pole_rate_by_nationality.plot(kind='bar', figsize=(10, 6))
plt.title('Top 10 Nationalities with the Highest Average Pole Rates')
plt.xlabel('Nationality')
plt.ylabel('Average Pole Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
#Top 10 Drivers with the Highest Points per Entry
```

```python
top_points_per_entry = df.nlargest(10, 'Points_Per_Entry')
top_points_per_entry.plot(kind='bar', x='Driver', y='Points_Per_Entry', figsize=(12, 6))
plt.title('Top 10 Drivers with the Highest Points per Entry')
plt.xlabel('Driver')
plt.ylabel('Points per Entry')
plt.xticks(rotation=45)
plt.show()
```

```python
#No. of drivers in each decade
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Decade'], bins=range(1950, 2031, 10), edgecolor='k', alpha=0.7)
plt.title('Number of Drivers in Each Decade')
plt.xlabel('Decade')
plt.ylabel('Number of Drivers')
plt.xticks(range(1950, 2031, 10), rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
